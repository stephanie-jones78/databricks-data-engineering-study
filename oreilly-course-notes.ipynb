## Lakehouse Platform
Data Warehouse > Data Lake > **Data Lakehouse**

>from Databricks website, What is a Data Lakehouse

Limitations of traditional **Data Warehouses** handling unstructured data at scale<br>
Lack of governance in **Data Lakes**<br>
**Data Lakehouses** combine the strengths of Data Warehouses and Data Lakes**
- Can work with *structured and unstructured data*
- *Metadata* and *governance* layer between source and destination
- *Flexibility* (BI, reporting, data science, and ML)
- *Separates* storage and compute functions
- *Simplified architecture* for different systems
- *Faster time to insight* (no need to constantly copy and move data)


## Databricks Architecture
Control Plane (managed by Databricks)
- Processes data and workload

Compute Plane (lives in customer cloud - AWS, Azure, GCP, etc.)
- Handles backend compute resources and processes 

## Databricks Clusters
Databricks clusters are **computing resources** you can set up in your Databricks workspace for various data tasks like ETL pipelines, analytics, and machine learning. 

You can manage these resources using Databricks UI, CLI, or REST API. Databricks solves the problem of managing complex data analytics infrastructure, offering scalable and flexible compute options like: 
- **all-purpose compute** for *interactive analysis*, 
- **job compute** for *automated tasks*, 
- **instance pools** for *efficient resource use*, and 
- **SQL warehouses** for *SQL tasks*

This flexibility provides organizations with opportunities to optimize their data processing, reduce costs, and improve data-driven decision-making.
